```python
import datetime
import pytz
from pytz import timezone

utc_time = datetime.datetime.utcnow()

#adds new column "adjusted_hours" to X with hour timestamps adjusted for local timezones
def create_adjusted_timestamp(x):
    adjusted_timestamps = []
    for row in x.iterrows():
        timezone = row[1][19]
        #if there is no timezone data then I'm just setting the default to 0 for now but no idea what to do
        if pd.isna(timezone):
            adjusted_timestamps.append(0)
        else:
            adjusted_timestamps.append(pytz.utc.localize(utc_time, is_dst=None)
                                       .astimezone(pytz.timezone(timezone)).hour)
    x["adjusted_hours"] = adjusted_timestamps
    
create_adjusted_timestamp(table_11_00)
table_11_00



```
